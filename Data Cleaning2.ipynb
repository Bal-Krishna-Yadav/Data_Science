```python
import pandas as pd

# Sample data
data = {
    'Name': ['Bal Krishna', 'Deepak Kumar', None,None, 'Sohan kumar', '', 'Soni kumari', 'Moni Kumari', 'Sakshi Kumari', 'Chandni Kumari',
             'Nisha Kumari'],
    'Age': [25, 30, 22, None,None, 28, 35, -5, 29, 30, 40,],
    'Gender': ['F', 'M', 'M', 'M', 'F', None,None, 'M', 'F', 'M', ''],
    'Salary': [50000, 60000, 52000, 58000, None,None, 62000, 0, 59000, 60000, 61000]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull().sum())
```

```python
df.shape
```

```python
df['Name'].fillna('Krishna',inplace=True)
```

```python
df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
```

```python
df
```

```python
df['Salary'].fillna(df['Salary'].median(), inplace=True)
```

```python
df
```

```python
# Remove Negative Age
df = df[df['Age'] > 0]
```

```python
df
```

```python
df['Name'] = df['Name'].str.title()
```

```python
df
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# Encode Categorical Data (if needed)
df['Gender'] = df['Gender'].map({'M': 0, 'F': 1})
```
